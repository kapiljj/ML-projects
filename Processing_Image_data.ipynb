Convert RGB Image to grayscale Image using OpenCV

```python
#importing opencv  library
import cv2
```

```python
img= cv2.imread('/content/0105b5a8865355f0c551606c4fee9120.jpg')

grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

```
